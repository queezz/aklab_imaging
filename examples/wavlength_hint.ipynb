```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

%matplotlib inline
```

# Rough calibration
Overall calibration to know roughly correspondence between motor position and central wavelength.

```python
# Data points
x = [632.4, 584, 399]
y = [434, 486, 656]

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Define the fit line
x_fit = np.linspace(min(x), max(x), 100)
y_fit = slope * x_fit + intercept

def lam_to_pos(lam):
    return (lam-intercept)/slope

# Plot the data and the fit line
plt.plot(x, y, 'o', label='Data points')
plt.plot(x_fit, y_fit, '-', label=f'Fit: y = {slope:.2f}x + {intercept:.2f}')
plt.xlabel('position/1000')
plt.ylabel('wavelength, nm')
plt.legend()
plt.title(f"Linear Regression (RÂ² = {r_value**2:.2f})")
plt.show()

# Print regression details
print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"R-squared: {r_value**2:.2f}")
```
